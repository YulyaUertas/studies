Файл moscow_places.csv:
* __name__ — название заведения;
* __address__ — адрес заведения;
* __category__ — категория заведения, например «кафе», «пиццерия» или «кофейня»;
* __hours__ — информация о днях и часах работы;
* __lat__ — широта географической точки, в которой находится заведение;
* __lng__ — долгота географической точки, в которой находится заведение;
* __rating__ — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);
* __price__ — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
* __avg_bill__ — строка, которая хранит среднюю стоимость заказа в виде диапазона, например:
  - «Средний счёт: 1000–1500 ₽»;
  - «Цена чашки капучино: 130–220 ₽»;
  - «Цена бокала пива: 400–600 ₽».и так далее;
* __middle_avg_bill__ — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»:
  - Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
  - Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
  - Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
* __middle_coffee_cup__ — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»:
  - Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
  - Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
  - Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.
* __chain__ — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
  - 0 — заведение не является сетевым
  - 1 — заведение является сетевым
* __district__ — административный район, в котором находится заведение, например Центральный административный округ;
* __seats__ — количество посадочных мест.

## Обработка данных

```python
# ищем полные дубликаты
print ('Количество дубликатов:',data.duplicated().sum() )
```

```python
# ищем все пропуски
data.isna().sum()
```

В нашем файле пропуски есть в следующих столбцах: hours, price, avg_bill, middle_avg_bill, middle_coffee_cup, seats. Мы не можем обработать пропуски ни в одном из столбов, так как каждое заведение, даже если оно сетевое, может иметь свой режим работы, свой прайс, количество посадочных мест и т.д. Мы заменим пропуски только в столбце с режимом работы. Поставим вместо пропуска заглушку "нет".

```python
# заменим пропус
data['hours']=data['hours'].fillna('нет')
```

```python
# добавляем столбец street с названиями улиц
# пишем функцию, которая будет разделять строчку адреса на части, 
# будет брать часть с улицей, удалять слово 'улица' и все пробелы
def find_street (stroc):
    stroc=stroc.split(',')[1]
    stroc=stroc.replace('улица','')
    stroc=stroc.strip()
    return stroc
# применяем функцию
data['street']=data['address'].apply(find_street)
```

```python
# добавляем столбец is_24/7, в котором будет отражаться True - если заведение работает ежедневно и круглосуточно
# и False - если нет
# пишем функцию, которая будет находить слова 'ежедневно' и 'круглосуточно' и присвать True, иначе False
def work_schedule (hours):
    if 'ежедневно' and 'круглосуточно' in hours:
        return 'True'
    if 'нет' in hours:
        return 'Нет'
    return 'False' 

# применеям функцию
data['is_24/7']=data['hours'].apply(work_schedule )
# проверяем, что получилось
data.head()
```

```python
# проверим, режим работы - ежедневно, круглосуточно
data.loc[data['is_24/7']=='True'].head()
```

Мы провели предработку данных. Проверили данные на дубликаты, их у нас нет. Все данные соотвествуют своим типам. Пропуски в данных есть, что в дальнейшем при анализе мы будем иметь ввиду. Так же мы добавили 2 новым столбца, первый - с названиями улиц, где находятся заведения, второй - с информацией по режиму работы, а именно, если заведение работает 24/7, то значение True, если нет False.

Далее посмотрим, как растояние от центра влияет на цену.

```python
# напишем функцию, которая посчитает растояние от центра Центрального округа до каждого заведения в километрах
def distance_2(x): 
    x['km']=[]  
    # Математический модуль содержит имя функции "радианы", которая используется для преобразования значения градусов в радианы
    Lng1 = radians(37.621184) 
    Lng2 = radians(x['lng']) 
    Lat1 = radians(55.753600) 
    Lat2 = radians(x['lat']) 
       
    # Используем формулу Хаверсина
    D_Lo = Lng2 - Lng1 
    D_La = Lat2 - Lat1 
    P = sin(D_La / 2)**2 + cos(Lat1) * cos(Lat2) * sin(D_Lo / 2)**2 
  
    Q = 2 * asin(sqrt(P)) 
     
    # Радиус Земли в километрах
    R_km = 6371 
    x['km']=round((Q * R_km))  
    return x
```

```python
# применяем функцию к датасету
data2 = data2.apply(lambda x: distance_2(x), axis=1)
data2.head()
```
